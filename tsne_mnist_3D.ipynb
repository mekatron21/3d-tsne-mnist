```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets, manifold

# Load the MNIST dataset
digits = datasets.load_digits()
X = digits.data
y = digits.target

# Perform t-SNE to reduce the dimensionality of the data to 3D
tsne = manifold.TSNE(n_components=3, init='pca')
X_tsne = tsne.fit_transform(X)

# Create a 3D scatter plot of the t-SNE results
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_tsne[:, 0], X_tsne[:, 1], X_tsne[:, 2], c=y, cmap=plt.cm.get_cmap('jet', 10))
ax.set_title('t-SNE visualization of MNIST dataset in 3D')
ax.set_xlabel('t-SNE dimension 1')
ax.set_ylabel('t-SNE dimension 2')
ax.set_zlabel('t-SNE dimension 3')
plt.show()
```
